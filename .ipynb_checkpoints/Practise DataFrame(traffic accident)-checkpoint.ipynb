```python
df = pd.read_csv('도로교통공단_전국_사망교통사고정보(2018).csv',encoding='euc-kr')
df
```

```python
# 1. 서울에서 일어난 요일별
# 사망자수, 사상자수 를 출력하시요
# 2. 서울 강북구 지역의 교통사고 위치를
# 지도로 표시하고 각마커에는 법규위반내용
# 을 표시하시요
# 3. 부산에서 발생한 총 사망자수를 구하시요
# 4. 서울지역 월요일에 발생한 사상자수
# 사고유형 법규위반을 구하시요
# 5.횡단중 일어난 사고의 요일 
# 발생지시도, 법규위반, 피해자_당사자종별
# 을 구하시요
```

```python
#1.
df[df.loc[:,'발생지시도']=='서울'].groupby(['요일'])[['사망자수','사상자수']].sum().loc[['월','화','수','목','금','토','일']]
```

```python
df[df['발생지시도']=='서울'][['요일','사망자수','중상자수']].groupby('요일').sum().loc[['월','화','수','목','금','토','일']]
```

```python
# 2. 서울 강북구 지역의 교통사고 위치를
# 지도로 표시하고 각마커에는 법규위반내용
# 을 표시하시요
Seoul_Gangbuk_Accidents = df[df['발생지시군구']=='강북구'][['법규위반','위도','경도']]

map = folium.Map(location=(37.63799365202427, 127.02543008253372),zoom_start=13)

for seoul_gangbuk_accident in Seoul_Gangbuk_Accidents.values:
    violation = seoul_gangbuk_accident[0]
    latitude = seoul_gangbuk_accident[1]
    longitude = seoul_gangbuk_accident[2]
    lat_long = (latitude,longitude)
    mk = folium.Marker(location=lat_long,popup=violation)
    mk.add_to(map)

map
```

```python
kangbuk =df[ (df['발생지시도']=='서울') & (df['발생지시군구']=='강북구')][['법규위반','위도','경도']]
kangbuk

map =folium.Map( location=(kangbuk['위도'].mean(),kangbuk['경도'].mean()),zoom_start=14 )
for i,s in kangbuk.iterrows():
    mk = folium.Marker( location=(s['위도'],s['경도']),popup=s['법규위반'] )
    mk.add_to(map)
map
map.save('kangbuk.html')
```

```python
# 3. 부산에서 발생한 총 사망자수를 구하시요
df[df['발생지시도']=='부산']['사망자수'].sum()
```

```python
# 4. 서울지역 월요일에 발생한 사상자수
# 사고유형 법규위반을 구하시요
df[(df['발생지시도']=='서울') & (df['요일']=='월')].loc[:,['사상자수','사고유형','법규위반']]
```

```python
# 5.횡단중 일어난 사고의 요일 
# 발생지시도, 법규위반, 피해자_당사자종별
# 을 구하시요
df.columns
df[df['사고유형']=='횡단중'][['요일','발생지시도','법규위반','피해자_당사자종별']]
```
